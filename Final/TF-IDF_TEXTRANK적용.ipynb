```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from tqdm import tqdm

# 형태소 분석기
okt = Okt()

# 텍스트 전처리 - 고유명사(NNP) 추출
def extract_proper_nouns(text):
    tokens_with_pos = okt.pos(text)
    proper_nouns = [word for word, pos in tokens_with_pos if pos == 'NNP']
    return ' '.join(proper_nouns)

# 텍스트 전처리 - 형용사 또는 형용어구 추출
def extract_adjective_phrases(text):
    tokens_with_pos = okt.pos(text)
    adjective_phrases = [word for word, pos in tokens_with_pos if pos == 'Adjective']
    return ' '.join(adjective_phrases)

# 키워드를 상위, 중위, 하위로 나누어 추출하는 함수
def extract_keywords_by_rank(tfidf_scores_dict, top_n=10):
    sorted_keywords = sorted(tfidf_scores_dict.items(), key=lambda x: x[1], reverse=True)
    highest_keywords = sorted_keywords[:top_n]
    mid_keywords = sorted_keywords[top_n:2 * top_n]
    lowest_keywords = sorted_keywords[2 * top_n:]
    return highest_keywords, mid_keywords, lowest_keywords

# 텍스트가 비어있거나 유효한 단어가 없는 경우를 걸러내는 함수
def remove_empty_texts(text_list):
    return [text for text in text_list if text.strip() != '']  # 공백을 제외한 유효한 텍스트만 반환

# 가게별 리뷰에서 키워드 추출 - TF-IDF 기반
def extract_keywords_from_reviews(df, content_column, top_n=10):
    # 리뷰 본문이 있는 리뷰만 추출 (비어있거나 공백인 리뷰는 제외)
    filtered_df = df[df[content_column].notnull() & df[content_column].str.strip().astype(bool)]
    
    if len(filtered_df) == 0:
        return None  # 유효한 데이터가 없으면 None 반환
    
    # 고유명사와 형용어구 추출
    proper_nouns_texts = filtered_df[content_column].apply(extract_proper_nouns).tolist()
    adjective_phrases_texts = filtered_df[content_column].apply(extract_adjective_phrases).tolist()

    proper_nouns_texts = remove_empty_texts(proper_nouns_texts)
    adjective_phrases_texts = remove_empty_texts(adjective_phrases_texts)

    # 유효한 텍스트가 없으면 None 반환
    if len(proper_nouns_texts) == 0 and len(adjective_phrases_texts) == 0:
        return None
    
    # 고유명사에 대한 TF-IDF 실행
    if len(proper_nouns_texts) > 0 and any(len(text.strip()) > 0 for text in proper_nouns_texts):
        vectorizer = TfidfVectorizer(max_features=min(100, len(proper_nouns_texts)))
        try:
            proper_nouns_matrix = vectorizer.fit_transform(proper_nouns_texts)
            proper_nouns_feature_names = vectorizer.get_feature_names_out()
            proper_nouns_tfidf_scores = proper_nouns_matrix.sum(axis=0).A1
            proper_nouns_scores_dict = dict(zip(proper_nouns_feature_names, proper_nouns_tfidf_scores))
            proper_nouns_high, proper_nouns_mid, proper_nouns_low = extract_keywords_by_rank(proper_nouns_scores_dict, top_n)
        except ValueError:
            proper_nouns_high, proper_nouns_mid, proper_nouns_low = [], [], []
    else:
        proper_nouns_high, proper_nouns_mid, proper_nouns_low = [], [], []

    # 형용어구에 대한 TF-IDF 실행
    if len(adjective_phrases_texts) > 0 and any(len(text.strip()) > 0 for text in adjective_phrases_texts):
        vectorizer = TfidfVectorizer(max_features=min(100, len(adjective_phrases_texts)))
        try:
            adjective_phrases_matrix = vectorizer.fit_transform(adjective_phrases_texts)
            adjective_phrases_feature_names = vectorizer.get_feature_names_out()
            adjective_phrases_tfidf_scores = adjective_phrases_matrix.sum(axis=0).A1
            adjective_phrases_scores_dict = dict(zip(adjective_phrases_feature_names, adjective_phrases_tfidf_scores))
            adjectives_high, adjectives_mid, adjectives_low = extract_keywords_by_rank(adjective_phrases_scores_dict, top_n)
        except ValueError:
            adjectives_high, adjectives_mid, adjectives_low = [], [], []
    else:
        adjectives_high, adjectives_mid, adjectives_low = [], [], []

    return (proper_nouns_high, adjectives_high), (proper_nouns_mid, adjectives_mid), (proper_nouns_low, adjectives_low)

# 데이터셋을 불러와 가게별로 리뷰 키워드 추출
def process_reviews_by_store(file_path, content_column='content', top_n=10):
    df = pd.read_csv(file_path, encoding='utf-8')
    
    # 가게별로 그룹화하여 리뷰 키워드를 추출
    store_keywords = defaultdict(dict)
    for store, group in tqdm(df.groupby('store_name'), desc=f"Processing reviews by store"):
        keywords = extract_keywords_from_reviews(group, content_column, top_n)
        if keywords is None:
            print(f"No valid data found for store {store}")
            continue
        
        # 키워드 저장
        high, mid, low = keywords
        proper_nouns_high, adjectives_high = high
        proper_nouns_mid, adjectives_mid = mid
        proper_nouns_low, adjectives_low = low

        store_keywords[store]['proper_nouns'] = {
            'high': proper_nouns_high,
            'mid': proper_nouns_mid,
            'low': proper_nouns_low
        }
        store_keywords[store]['adjectives'] = {
            'high': adjectives_high,
            'mid': adjectives_mid,
            'low': adjectives_low
        }
    
    return store_keywords

# 가게별 리뷰 키워드 추출 실행
file_path = './UJ_review_summary/데이터셋/경양식_레이블utf8.csv'
store_keywords = process_reviews_by_store(file_path, content_column='content', top_n=10)

# 결과 출력 (예시)
for store, keywords in store_keywords.items():
    print(f"\nStore: {store}")
    print(f"  Highest Proper Nouns: {keywords['proper_nouns']['high']}")
    print(f"  Highest Adjective Phrases: {keywords['adjectives']['high']}")
    print(f"  Mid Proper Nouns: {keywords['proper_nouns']['mid']}")
    print(f"  Mid Adjective Phrases: {keywords['adjectives']['mid']}")
    print(f"  Lowest Proper Nouns: {keywords['proper_nouns']['low']}")
    print(f"  Lowest Adjective Phrases: {keywords['adjectives']['low']}")
```

```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from tqdm import tqdm

# 형태소 분석기
okt = Okt()

# 텍스트 전처리 - 고유명사(NNP) 추출
def extract_proper_nouns(text):
    tokens_with_pos = okt.pos(text)
    proper_nouns = [word for word, pos in tokens_with_pos if pos == 'NNP']
    return ' '.join(proper_nouns)

# 텍스트 전처리 - 형용사 또는 형용어구 추출
def extract_adjective_phrases(text):
    tokens_with_pos = okt.pos(text)
    adjective_phrases = [word for word, pos in tokens_with_pos if pos == 'Adjective']
    return ' '.join(adjective_phrases)

# 키워드를 상위, 중위, 하위로 나누어 추출하는 함수
def extract_keywords_by_rank(tfidf_scores_dict, top_n=10):
    sorted_keywords = sorted(tfidf_scores_dict.items(), key=lambda x: x[1], reverse=True)
    highest_keywords = sorted_keywords[:top_n]
    mid_keywords = sorted_keywords[top_n:2 * top_n]
    lowest_keywords = sorted_keywords[2 * top_n:]
    return highest_keywords, mid_keywords, lowest_keywords

# 텍스트가 비어있거나 유효한 단어가 없는 경우를 걸러내는 함수
def remove_empty_texts(text_list):
    return [text for text in text_list if text.strip() != '']  # 공백을 제외한 유효한 텍스트만 반환

# 가게별 리뷰에서 키워드 추출 - TF-IDF 기반
def extract_keywords_from_reviews(df, content_column, top_n=10):
    # 리뷰 본문이 있는 리뷰만 추출 (비어있거나 공백인 리뷰는 제외)
    filtered_df = df[df[content_column].notnull() & df[content_column].str.strip().astype(bool)]
    
    if len(filtered_df) == 0:
        return None  # 유효한 데이터가 없으면 None 반환
    
    # 고유명사와 형용어구 추출
    proper_nouns_texts = filtered_df[content_column].apply(extract_proper_nouns).tolist()
    adjective_phrases_texts = filtered_df[content_column].apply(extract_adjective_phrases).tolist()

    proper_nouns_texts = remove_empty_texts(proper_nouns_texts)
    adjective_phrases_texts = remove_empty_texts(adjective_phrases_texts)

    # 유효한 텍스트가 없으면 None 반환
    if len(proper_nouns_texts) == 0 and len(adjective_phrases_texts) == 0:
        return None
    
    # 고유명사에 대한 TF-IDF 실행
    if len(proper_nouns_texts) > 0 and any(len(text.strip()) > 0 for text in proper_nouns_texts):
        vectorizer = TfidfVectorizer(max_features=min(100, len(proper_nouns_texts)))
        try:
            proper_nouns_matrix = vectorizer.fit_transform(proper_nouns_texts)
            proper_nouns_feature_names = vectorizer.get_feature_names_out()
            proper_nouns_tfidf_scores = proper_nouns_matrix.sum(axis=0).A1
            proper_nouns_scores_dict = dict(zip(proper_nouns_feature_names, proper_nouns_tfidf_scores))
            proper_nouns_high, proper_nouns_mid, proper_nouns_low = extract_keywords_by_rank(proper_nouns_scores_dict, top_n)
        except ValueError:
            proper_nouns_high, proper_nouns_mid, proper_nouns_low = [], [], []
    else:
        proper_nouns_high, proper_nouns_mid, proper_nouns_low = [], [], []

    # 형용어구에 대한 TF-IDF 실행
    if len(adjective_phrases_texts) > 0 and any(len(text.strip()) > 0 for text in adjective_phrases_texts):
        vectorizer = TfidfVectorizer(max_features=min(100, len(adjective_phrases_texts)))
        try:
            adjective_phrases_matrix = vectorizer.fit_transform(adjective_phrases_texts)
            adjective_phrases_feature_names = vectorizer.get_feature_names_out()
            adjective_phrases_tfidf_scores = adjective_phrases_matrix.sum(axis=0).A1
            adjective_phrases_scores_dict = dict(zip(adjective_phrases_feature_names, adjective_phrases_tfidf_scores))
            adjectives_high, adjectives_mid, adjectives_low = extract_keywords_by_rank(adjective_phrases_scores_dict, top_n)
        except ValueError:
            adjectives_high, adjectives_mid, adjectives_low = [], [], []
    else:
        adjectives_high, adjectives_mid, adjectives_low = [], [], []

    return (proper_nouns_high, adjectives_high), (proper_nouns_mid, adjectives_mid), (proper_nouns_low, adjectives_low)

# 데이터셋을 불러와 가게별로 리뷰 키워드 추출
def process_reviews_by_store(file_path, content_column='content', top_n=10):
    df = pd.read_csv(file_path, encoding='utf-8')
    
    # 가게별로 그룹화하여 리뷰 키워드를 추출
    store_keywords = defaultdict(dict)
    for store, group in tqdm(df.groupby('store_name'), desc=f"Processing reviews by store"):
        keywords = extract_keywords_from_reviews(group, content_column, top_n)
        if keywords is None:
            print(f"No valid data found for store {store}")
            continue
        
        # 키워드 저장
        high, mid, low = keywords
        proper_nouns_high, adjectives_high = high
        proper_nouns_mid, adjectives_mid = mid
        proper_nouns_low, adjectives_low = low

        store_keywords[store]['proper_nouns'] = {
            'high': proper_nouns_high,
            'mid': proper_nouns_mid,
            'low': proper_nouns_low
        }
        store_keywords[store]['adjectives'] = {
            'high': adjectives_high,
            'mid': adjectives_mid,
            'low': adjectives_low
        }
    
    return store_keywords

# 가게별 리뷰 키워드 추출 실행
file_path = './UJ_review_summary/데이터셋/경양식_레이블utf8.csv'
store_keywords = process_reviews_by_store(file_path, content_column='content', top_n=10)

# 가게별 키워드를 CSV로 저장하는 함수
def save_store_keywords_to_csv(store_keywords, output_dir='./output'):
    # 각 가게별로 키워드를 저장
    for store, keywords in store_keywords.items():
        # 고유명사 데이터프레임 생성
        proper_nouns_df = pd.DataFrame({
            'Highest Proper Nouns': [k for k, _ in keywords['proper_nouns']['high']],
            'Mid Proper Nouns': [k for k, _ in keywords['proper_nouns']['mid']],
            'Lowest Proper Nouns': [k for k, _ in keywords['proper_nouns']['low']]
        })

        # 형용어구 데이터프레임 생성
        adjectives_df = pd.DataFrame({
            'Highest Adjective Phrases': [k for k, _ in keywords['adjectives']['high']],
            'Mid Adjective Phrases': [k for k, _ in keywords['adjectives']['mid']],
            'Lowest Adjective Phrases': [k for k, _ in keywords['adjectives']['low']]
        })

        # 각각의 가게에 대한 CSV 파일로 저장
        proper_nouns_df.to_csv(f'{output_dir}/{store}_proper_nouns.csv', encoding='utf-8-sig', index=False)
        adjectives_df.to_csv(f'{output_dir}/{store}_adjective_phrases.csv', encoding='utf-8-sig', index=False)

# 저장할 파일 경로 지정
output_dir = './UJ_review_summary/결과'
save_store_keywords_to_csv(store_keywords, output_dir)

```
