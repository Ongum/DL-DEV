```python
import pandas as pd
import torch
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertTokenizer, BertForSequenceClassification
from tqdm import tqdm
from soynlp.tokenizer import LTokenizer
from gensim.models import LdaModel
from gensim.corpora import Dictionary
```

```python
# 1. 데이터 로드 및 전처리
file_path = './외국음식전문점.csv'
df = pd.read_csv(file_path)
```

```python
# 리뷰 컬럼 추출 (컬럼 이름을 실제 데이터에 맞게 변경)
text_data = df['content'].dropna().tolist()
```

```python
# Soynlp LTokenizer 사용
from soynlp.word import WordExtractor
```

```python
# WordExtractor로 단어 점수 학습
word_extractor = WordExtractor()
word_extractor.train(text_data)
word_scores = word_extractor.extract()
```

```python
# word_scores에서 Scores 객체를 float 값으로 변환
float_word_scores = {word: score.cohesion_forward for word, score in word_scores.items()}
```

```python
# LTokenizer를 사용해 텍스트 토큰화
tokenizer = LTokenizer(scores=float_word_scores)
```

```python
# tqdm을 사용한 리뷰 데이터 토큰화
tokenized_data = []
for review in tqdm(text_data, desc="Tokenizing Reviews with Soynlp"):
    tokens = tokenizer.tokenize(review)
    tokenized_data.append(tokens)
    
    print("Tokenization completed successfully.")
```

```python
# 2. LDA 모델을 사용한 토픽 모델링
# Dictionary 및 Corpus 생성
dictionary = Dictionary(tokenized_data)
corpus = [dictionary.doc2bow(text) for text in tokenized_data]
```

```python
# LDA 모델 학습을 위한 데이터 준비
dictionary = Dictionary(tokenized_data)
corpus = [dictionary.doc2bow(text) for text in tokenized_data]

# LDA 모델 학습
num_topics = 10  # 토픽 수 설정
passes = 15  # 학습 패스 수 설정

# tqdm으로 학습 상황 전개
pbar = tqdm(range(passes), desc="LDA Model Training")

lda_model = LdaModel(
    corpus=corpus, 
    num_topics=num_topics, 
    id2word=dictionary, 
    passes=1  # 한 번에 하나의 pass만 수행하도록 설정
)

for _ in pbar:
    lda_model.update(corpus)  # 한 번의 pass에 대해 모델 업데이트
    pbar.set_postfix({"Current Pass": _ + 1})

print("LDA Model Training Completed.")
```

```python
# 토픽 출력
topics = lda_model.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
# 3. KoBERT를 사용한 감성 분석
# GPU 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device set to: {device}")
```

```python
# KoBERT 모델 및 토크나이저 로드
tokenizer_bert = BertTokenizer.from_pretrained('monologg/kobert')
model = BertForSequenceClassification.from_pretrained('monologg/kobert', num_labels=2)
model.to(device)
```

```python
# 텍스트 데이터 토큰화 및 데이터셋 생성
inputs = tokenizer_bert(text_data, return_tensors='pt', padding=True, truncation=True, max_length=128)
labels = torch.tensor([0] * len(text_data))  # 예시로 모든 레이블을 0으로 설정 (실제 감성 레이블 필요)
dataset = TensorDataset(inputs['input_ids'], inputs['attention_mask'], labels)
train_loader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
# 옵티마이저 및 손실 함수 설정
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
if torch.cuda.is_available():
    print("CUDA is available. Training on GPU.")
else:
    print("CUDA is not available. Training on CPU.")
```

```python
# KoBERT 학습 루프
epochs = 3
model.to(device)

for epoch in range(epochs):
    model.train()
    epoch_loss = 0
    for batch_idx, batch in enumerate(tqdm(train_loader, desc=f"KoBERT Training Epoch {epoch+1}/{epochs}")):
        input_ids, attention_mask, labels = [x.to(device) for x in batch]
        
        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        
        loss = loss_fn(logits, labels)
        epoch_loss += loss.item()
        
        loss.backward()
        optimizer.step()

        # 10개 배치마다 출력
        if batch_idx % 100 == 0:
            tqdm.write(f"Batch {batch_idx} Loss: {loss.item():.4f}")
    
    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss/len(train_loader):.4f}")
```

```python
# 4. KC-BERT를 사용한 신조어 처리 및 감성 분석
# KC-BERT 모델 및 토크나이저 로드
tokenizer_kcbert = BertTokenizer.from_pretrained('beomi/kcbert-base')
model_kcbert = BertForSequenceClassification.from_pretrained('beomi/kcbert-base', num_labels=2)
model_kcbert.to(device)
```

```python
# KC-BERT 텍스트 데이터 토큰화 및 데이터셋 생성
inputs_kcbert = tokenizer_kcbert(text_data, return_tensors='pt', padding=True, truncation=True, max_length=128)
```

```python
# 레이블을 입력 데이터 크기에 맞게 재생성
labels = torch.tensor([0, 1] * (inputs_kcbert['input_ids'].size(0) // 2 + 1))[:inputs_kcbert['input_ids'].size(0)]

# 데이터셋 생성
dataset_kcbert = TensorDataset(inputs_kcbert['input_ids'], inputs_kcbert['attention_mask'], labels)
train_loader_kcbert = DataLoader(dataset_kcbert, batch_size=16, shuffle=True)
```

```python

# KC-BERT 학습 루프
for epoch in range(epochs):
    model_kcbert.train()
    epoch_loss_kcbert = 0
    for batch in tqdm(train_loader_kcbert, desc=f"KC-BERT Training Epoch {epoch+1}/{epochs}"):
        input_ids, attention_mask, labels = [x.to(device) for x in batch]
        optimizer.zero_grad()
        outputs = model_kcbert(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        loss = loss_fn(logits, labels)
        epoch_loss_kcbert += loss.item()
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss_kcbert/len(train_loader_kcbert)}")
```

```python
# 데이터프레임에 가게 이름이 포함된 컬럼이 있다고 가정 ('store_name')
grouped_reviews = df.groupby('store_name')['content'].apply(lambda x: ' '.join(x)).reset_index()

# 각 가게별 리뷰를 저장할 리스트
store_summaries = []
```

```python
bart_tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
bart_model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn').to(device)  # 모델을 GPU로 이동

# 데이터 예시
store_name = "Example Store"
combined_reviews = "이 가게는 정말 훌륭합니다. 음식이 매우 맛있고, 서비스가 최고입니다."  # 예시 리뷰

# 토큰화
inputs = bart_tokenizer(combined_reviews, return_tensors='pt', max_length=1024, truncation=True)
print(f"Inputs device: {inputs['input_ids'].device}")  # CPU일 경우
```

```python
# 입력 데이터를 모델과 동일한 디바이스로 이동
inputs = {key: tensor.to(device) for key, tensor in inputs.items()}
print(f"Inputs device after moving to device: {inputs['input_ids'].device}")  # GPU로 이동 후 디바이스 확인
```

```python
summary_ids = bart_model.generate(
    inputs['input_ids'], 
    attention_mask=inputs['attention_mask'],  # 명시적으로 attention mask 추가
    max_length=150, 
    min_length=40, 
    length_penalty=2.0, 
    num_beams=4, 
    early_stopping=True
)

summary = bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(f"Summary: {summary}")
```

```python
bart_model.eval()
with torch.no_grad():
    for index, row in tqdm(grouped_reviews.iterrows(), desc="Summarizing Reviews by Store", total=len(grouped_reviews)):
        store_name = row['store_name']
        combined_reviews = row['content']
        
        # 가게에 대한 모든 리뷰를 하나의 입력으로 요약
        inputs = bart_tokenizer(combined_reviews, return_tensors='pt', max_length=1024, truncation=True).to(device)
        
        # BART 모델의 입력 텐서를 모델이 있는 디바이스로 이동
        inputs = {key: tensor.to(device) for key, tensor in inputs.items()}
        
        # 모델로 요약 생성
        summary_ids = bart_model.generate(
            inputs['input_ids'], 
            attention_mask=inputs['attention_mask'],  # 이 줄을 추가합니다.
            max_length=150, 
            min_length=40, 
            length_penalty=2.0, 
            num_beams=4, 
            early_stopping=True
        )
        summary = bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        
        store_summaries.append({'store_name': store_name, 'summary': summary})
        
        # 콘솔에 요약 결과 출력
        print(f"Store: {store_name}\nSummary: {summary}\n")

# 가게별 요약 결과를 파일로 저장
store_summaries_df = pd.DataFrame(store_summaries)
store_summaries_df.to_csv("store_review_summaries.csv", index=False)
print("Store review summaries saved to store_review_summaries.csv")
```
